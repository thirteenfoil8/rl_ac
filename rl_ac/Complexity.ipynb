```python
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
```

```python
M = np.ones((4,3))
labels = ['2D Game','Rocket Master','Gran Tourismo','Assetto Corsa']
```

```python
M[0][0]=10
M[0][1]=11
M[0][2]=12
M[1][0]=10
M[1][1]=30
M[1][2]=40
M[2][0]=10
M[2][1]=50
M[2][2]=80
M[3][0]=20
M[3][1]=80
M[3][2]=160
```

```python
M=np.transpose(M)
M
```

```python
fig, ax = plt.subplots()
width = 0.35
ax.bar(labels, M[0], width,  label='10Kmh')
ax.bar(labels, M[1], width,bottom= M[0],  label='50Kmh')
ax.bar(labels, M[2], width,bottom=M[1]+M[0],  label='100Kmh')


ax.set_ylabel('Possible next states')
ax.set_title('Possible next states by complexity of the simulation')
ax.legend()
# idea about stability has to be improved
```

```python
fig, ax = plt.subplots()
width = 0.35
x = np.arange(len(labels))
rects1 = ax.bar(x-width/1.5, M[0], width,  label='10Kmh')
rects2 = ax.bar(x, M[1], width,  label='50Kmh')
rects3 = ax.bar(x+width/1.5, M[2], width/1.25,  label='100Kmh')

plt.xticks(np.arange(0, 3, step=1))  # Set label locations.
plt.xticks(np.arange(0, 4, step=1), ['2D Game','Rocket Master','Gran Tourismo','Assetto Corsa'])  # Set text labels.

ax.set_ylabel('States Complexity')
ax.set_title('Next states complexity')
ax.legend()
# idea about stability has to be improved
```

```python
values  = np.array([1,1,1,1])
err = np.array([1,0.5,0.1,0.05])
```
